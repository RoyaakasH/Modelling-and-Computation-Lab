```python
import matplotlib.pyplot as plt
plot=sns.barplot(x="mpg",y="displacement",data=df)

x=plot.patches[3].get_x() + plot.patches[3].get_width() / 2
y=plot.lines[3].get_ydata()[1]+0.3
plt.annotate('Highest bar', xy=(x,y),xytext=(x+0.9,y+55), arrowprops=dict(facecolor='indigo'))
```

**Observations**

The displacement decreases as the mpg increases.

The highest value of displacement is 455.

$$Q:4$$

```python
import numpy as np
```

```python
plot_2=sns.barplot(x="mpg",y="displacement",hue='model_year',data=df,estimator=np.median)
plot_2.legend(loc='upper right',ncol=3)
```

**Observations:**

Here, the no. of mpg is very large. That's why we are getting this in white colour.

Displacement first started decreasing when mpg started increasing but thereafter, it became constant.

Adding annotations to denote the median values at the top boundary of the highest bars in the bar
plots.

```python
plot_2 = sns.barplot(x="mpg",y="displacement",hue="model_year",data=df,estimator=np.median)
for p,line in zip(plot_2.patches, plot_2.lines):
    plot_2.annotate(format(p.get_height(),'.2f'), 
                   (p.get_x() + p.get_width() / 2, line.get_ydata()[1]), 
                   ha = 'center', va = 'top', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plot_2.legend(loc='upper right',ncol=4)
```

```python
n=0
for i in range(len(df['displacement'])):
  if df['displacement'][i]==455:
    n=n+1
    j=i
print('The bar with mpg value',df['mpg'][j],' have the highest displacement.')
```

```python
plot_2 = sns.barplot(x="mpg",y="displacement",hue="model_year",data=df,estimator=np.median)
plot_2.annotate(np.median(df['mpg'][95]),xy=(df['mpg'][95], 455),xytext = (95, 455))
plot_2.legend(loc='upper right',ncol=4)
x=plot_2.patches[3].get_x() + plot_2.patches[3].get_width() / 2
y=plot_2.lines[1].get_ydata()[1]+0.3
plt.annotate('Highest bar value = 455.0', xy=(x,y),xytext=(x+0.9,y+55), arrowprops=dict(facecolor='indigo'))
```
